**1. Transform raw data into visual formats like charts, graphs, and dashboards.**

Feature Engineering (Add average score)

```python
# Create a new column for average score
df["average score"] = df[["math score", "reading score", "writing score"]].mean(axis=1)
```

Visualizations Using Seaborn & Matplotlib

*Histogram: Distribution of Average Score*

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.histplot(df["average score"], kde=True, bins=20, color="skyblue")
plt.title("Distribution of Average Student Scores")
plt.xlabel("Average Score")
plt.ylabel("Frequency")
plt.show()
```

*Bar Plot: Average Score by Gender*

```python
plt.figure(figsize=(6, 5))
sns.barplot(x="gender", y="average score", data=df, palette="pastel", ci=None)
plt.title("Average Score by Gender")
plt.ylabel("Average Score")
plt.show()
```

*Box Plot: Score Spread by Test Preparation*

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x="test preparation course", y="average score", data=df, palette="Set2")
plt.title("Score Distribution by Test Preparation Course")
plt.show()
```

*Heatmap: Correlation Between Scores*

```python
plt.figure(figsize=(6, 5))
sns.heatmap(df[["math score", "reading score", "writing score", "average score"]].corr(), annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap of Scores")
plt.show()
```

*Scatter Plot: Math vs Reading Score Colored by Gender*

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(data=df, x="math score", y="reading score", hue="gender", palette="coolwarm")
plt.title("Math vs Reading Score by Gender")
plt.show()
```

```python
import dash
from dash import dcc, html
import plotly.express as px
```

```python
# Load and process data
df["average score"] = df[["math score", "reading score", "writing score"]].mean(axis=1)
```

```python
# Start Dash app
app = dash.Dash(__name__)
app.title = "Student Performance Dashboard"
```

```python
# Layout
app.layout = html.Div([
    html.H1(" Student Performance Dashboard", style={'textAlign': 'center'}),

    dcc.Dropdown(
        id='group_by',
        options=[
            {'label': 'Gender', 'value': 'gender'},
            {'label': 'Lunch Type', 'value': 'lunch'},
            {'label': 'Test Prep Course', 'value': 'test preparation course'},
            {'label': 'Parental Education', 'value': 'parental level of education'},
        ],
        value='gender',
        clearable=False,
        style={'width': '50%', 'margin': 'auto'}
    ),

    dcc.Graph(id='bar_plot'),

    dcc.Graph(
        figure=px.box(df, x='gender', y='average score', color='gender',
                      title="Boxplot: Average Score by Gender")
    ),

    dcc.Graph(
        figure=px.scatter_matrix(df,
                                 dimensions=["math score", "reading score", "writing score"],
                                 color="gender",
                                 title="Scatter Matrix: Score Relationships")
    )
])
```

```python
# Callback to update bar plot
@app.callback(
    dash.dependencies.Output('bar_plot', 'figure'),
    [dash.dependencies.Input('group_by', 'value')]
)
def update_bar_chart(group_col):
    grouped = df.groupby(group_col)["average score"].mean().reset_index()
    fig = px.bar(grouped, x=group_col, y="average score", color=group_col,
                 title=f"Average Score by {group_col.title()}")
    return fig
```

```python
# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```
